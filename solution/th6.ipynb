# a

$\frac{dS}{dt}/\frac{dR}{dt}=\frac{dS}{dR}=-\beta S \rightarrow S=S_0 e^{-\frac{\beta}{\gamma} R}$

$1=I+S+R→I=1-R-S_0e^{-\frac{Β}{γ}R} , I_{∞}=0 →R=1-S_0e^{-\frac{Β}{γ}R}$

# b

```python
beta=0.4
gamma=0.2
R0=beta/gamma
t = np.linspace(0,200,num=200000)
I_0=0.1
S_0=1-I_0
R_0=0


def equations(y, t, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I 
    dIdt = beta * S * I - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt 
y0 =S_0,I_0,R_0
RES = sp.odeint(equations, y0, t, args= (beta, gamma))
S, I, R = RES.T
  

plt.figure(num=None, figsize=(6, 4), dpi=200, facecolor='w', edgecolor='k')
plt.plot(t,R,label='R')
plt.plot(t,I,label='I')
plt.plot(t,S,label='S')
plt.plot(t,S_0*np.exp(-beta/gamma*R),label='not sim')
plt.plot(t,1-R-S_0*np.exp(-beta/gamma*R),label='not sim')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Fraction of population')
```

## c

```python
gamma=0.2
Beta= np.linspace(0.5,5,num=100)*gamma
R_inf=[]
for beta in Beta:

  t = np.linspace(0,10000,num=200000)
  I_0=0.1
  S_0=1-I_0


  def equations(y, t, beta, gamma):
      S, I, R = y
      dSdt = -beta * S * I 
      dIdt = beta * S * I - gamma * I
      dRdt = gamma * I
      return dSdt, dIdt, dRdt 
  y0 =S_0,I_0,R_0
  RES = sp.odeint(equations, y0, t, args= (beta, gamma))
  S, I, R = RES.T
  R_inf.append(R[-1])
  

plt.figure(num=None, figsize=(6, 4), dpi=100, facecolor='w', edgecolor='k')
plt.plot(Beta/gamma,R_inf,label='R')
plt.xlabel(r'$\frac{\beta}{\gamma}$')
plt.ylabel(r'$R_\infty$')
```
